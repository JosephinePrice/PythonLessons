```python
final_urls = []
for x in new_urls:
    if len(x.split(',')) > 1:
        new_strings = [x.lstrip() for x in new_urls[8].split(',')]
        for item in new_strings:
            final_urls.append(item)
    else:
        final_urls.append(x)
final_urls
```

```python
for count, url in enumerate(final_urls):

    response = requests.get(url)
    if response.status_code == 200:
        with open("C:\\Users\\Winadmin\\Desktop\\taliban_images\\{}.jpg".format(count + 1), 'wb') as f:
            f.write(response.content)
    else:
        print("Error: {}".format(response.status_code))
        print("This happend for url: {}".format(url))
```
